En el analisis del dataSet me di cuenta que el año  2019 hay varias filas que tiene un *  y existe tambien un año con solo un * , este ultimo corresponde a la fila 864 la cual será eliminada por no tener ningun tipo de información.

```python
filas_con_asterisco = df1[df1['Año'] == '*']

# Imprimir las filas
print(filas_con_asterisco)
```

Vamos a modificar el el año '2019*' por '2019'   y eliminar la fila 864


```python
# Modificar las filas en la columna
df1['Año'] = df1['Año'].replace('2019 *', '2019')

#eliminar la fila 864 , ya que no tiene ningun valor .
df1 = df1.drop(864)
```

```python
valores_unicos = df1['Trimestre'].unique()
print(valores_unicos)
```

Aca onservamos que hay valores en el trimestre que tienen un *. Vamos hacer le reemplazo de estos valores, en  cada caso: 3 * po 3, 2 * por 2, 1 * por 1.

```python
df1['Trimestre'] = df1['Trimestre'].replace('3 *','3')
df1['Trimestre'] = df1['Trimestre'].replace('2 *','2')
df1['Trimestre'] = df1['Trimestre'].replace('1 *','1')
```

```python
df1.head()
```

```python

print(df1['Trimestre'].value_counts())
```

```python
df1.drop(df1.tail(1).index, inplace=True)
```

```python
df1[['Trimestre','Año']] = df1[['Trimestre','Año']] .astype(int)
     
```
