Estudio completo por  cada servicio. Aca vamos a estudiar el comportamiento de la penetración de la telefonía movil desde el 2014 hasta el 2022


1.Movil

```python
penetracion_movil= pd.read_csv('Telefonia_Movil/Penetración_nacional_de_la_telefonía_móvil(accesos por cada 100 habitantes).csv',index_col=0)
print(penetracion_movil.info())
penetracion_movil.head()
```

Calculamos el promedio de acceso por  año usando groupby.(movil)

```python
accesos_por_anio =penetracion_movil.groupby('Año')['Accesos por cada 100 hab'].sum()
print(accesos_por_anio)
```

Calculamos el acceso promedio por año de telefonia movil.

```python
accesos_prom_por_anio =penetracion_movil.groupby('Año')['Accesos por cada 100 hab'].mean()
print(accesos_prom_por_anio)
```

```python
# Graficar los accesos por año
plt.figure(figsize=(10, 6))
plt.bar(accesos_por_anio.index, accesos_por_anio.values, color='skyblue', label='Accesos por año')
plt.xlabel('Año')
plt.ylabel('Accesos por cada 100 hab')
plt.title('Accesos por año telefonía móvil')
plt.legend()
plt.grid(True)
plt.show()
```

Conclusiones con respecto a la penetración de la telefonia movil desde le 2014 hasta el 2022. Podemos  decir que   la telefonia movil ha mantenido una penetración constante a traves de los años y ha ido en un decrecimiento muy leve.
